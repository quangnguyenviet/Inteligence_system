```python
# in ra các giá trị khác nhau trong cột
import pandas as pd

df = pd.read_csv("DATA/weather_cleaned.csv")

# Đếm số giá trị khác nhau
unique_count = df["province"].nunique()

# Lấy danh sách các province khác nhau
unique_provinces = df["province"].unique()

print("Số lượng province khác nhau:", unique_count)
print("Danh sách province:")
for p in unique_provinces:
    print("-", p)
```
