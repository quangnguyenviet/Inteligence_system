```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

directory = r"C:\DATA\flower"
img_size = (150, 150)
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255, # change range (0, 1)
    rotation_range=30, # augumentation
    zoom_range=0.2, # augumentation
    horizontal_flip=True, # augumentation
    validation_split=0.2  # 20% validation
)

train_data = train_datagen.flow_from_directory(
    directory,
    target_size=img_size, # resize image
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_data = train_datagen.flow_from_directory(
    directory,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(5, activation='softmax')  # 5 classes
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
)

import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train acc')
plt.plot(history.history['val_accuracy'], label='Val acc')
plt.legend()
plt.show()


import numpy as np
from tensorflow.keras.preprocessing import image

img_path = "test_flower.jpg"
img = image.load_img(img_path, target_size=(150,150))
x = image.img_to_array(img) / 255.0
x = np.expand_dims(x, axis=0)

pred = model.predict(x)
classes = list(train_data.class_indices.keys())
print("LoÃ i hoa dá»± Ä‘oÃ¡n:", classes[np.argmax(pred)])

```

```python
import pandas as pd
import nlpaug.augmenter.word as naw
from googletrans import Translator
import random

# Äá»c dá»¯ liá»‡u
df = pd.read_csv("comments.csv")

# Khá»Ÿi táº¡o translator
translator = Translator()

# Back Translation (dá»‹ch Viá»‡t â†’ Anh â†’ Viá»‡t)
def back_translate(text):
    try:
        translated = translator.translate(text, src='vi', dest='en').text
        back_translated = translator.translate(translated, src='en', dest='vi').text
        return back_translated
    except:
        return text  # fallback náº¿u lá»—i

# Synonym Replacement (Tiáº¿ng Viá»‡t - Ä‘Æ¡n giáº£n)
vi_synonyms = {
    "tá»‘t": ["tuyá»‡t vá»i", "á»•n", "hÃ i lÃ²ng"],
    "tá»‡": ["kÃ©m", "tá»“i", "kinh khá»§ng"],
    "nhanh": ["mau", "chÃ³ng", "tá»©c thÃ¬"],
    "Ä‘Ã³ng gÃ³i": ["gÃ³i hÃ ng", "Ä‘Ã³ng kiá»‡n", "bao bá»c"]
}

def replace_synonyms(text):
    for word, syns in vi_synonyms.items():
        if word in text:
            text = text.replace(word, random.choice(syns))
    return text

# Táº¡o cá»™t má»›i chá»©a vÄƒn báº£n Ä‘Æ°á»£c augment
df['text_backtranslate'] = df['text'].apply(back_translate)
df['text_synonym'] = df['text'].apply(replace_synonyms)

# Gá»™p láº¡i vá»›i dá»¯ liá»‡u gá»‘c
df_augmented = pd.concat([
    df[['text', 'label']].rename(columns={'text': 'aug_text'}),
    df[['text_backtranslate', 'label']].rename(columns={'text_backtranslate': 'aug_text'}),
    df[['text_synonym', 'label']].rename(columns={'text_synonym': 'aug_text'})
], ignore_index=True)

# Xuáº¥t dá»¯ liá»‡u
df_augmented.to_csv("augmented_comments.csv", index=False)

print(df_augmented.head(10))
```
