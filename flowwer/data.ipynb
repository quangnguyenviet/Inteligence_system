```python
import pandas as pd
import numpy as np

N_USERS = 1000
N_ITEMS = 500
REVIEWS_PER_USER = 50

# Tạo User và Item IDs
user_ids = [f'User_{i:04d}' for i in range(1, N_USERS + 1)]
item_ids = [f'Item_{i:04d}' for i in range(1, N_ITEMS + 1)]

# Tạo dữ liệu ngẫu nhiên
data = []
for user in user_ids:
    # Mỗi user đánh giá 50 items ngẫu nhiên
    sampled_items = np.random.choice(item_ids, REVIEWS_PER_USER, replace=False)
    for item in sampled_items:
        rating = np.random.randint(1, 6) # Rating 1-5

        # Ví dụ Review tiếng Việt (thực tế cần phức tạp hơn)
        if rating >= 4:
            review = "Sản phẩm này rất tốt, giao hàng nhanh và đóng gói cẩn thận. Tôi rất hài lòng với chất lượng!"
        elif rating == 3:
            review = "Chất lượng ổn, không có gì đặc biệt. Màu sắc không được như mong đợi."
        else:
            review = "Thật thất vọng, sản phẩm lỗi và không dùng được. Đánh giá 1 sao."

        data.append([user, item, rating, review])

df = pd.DataFrame(data, columns=['UserID', 'ItemID', 'Rating', 'Review_VN'])

# Chuẩn bị tên file theo yêu cầu: itemReview_lớp.nhóm_tênho.csv
file_name = "data-ecom_quangnv.csv"
df.to_csv(file_name, index=False)
```

```python
import pandas as pd
import numpy as np
import random

# -------------------------------
# 1. Cấu hình cơ bản
# -------------------------------
N_USERS = 1000
REVIEWS_PER_USER = 50
CATEGORIES = ["phone", "laptop", "clothes"]

# Sinh danh sách người dùng
user_ids = [f'User_{i:04d}' for i in range(1, N_USERS + 1)]

data = []

# -------------------------------
# 2. Review mẫu cho từng loại
# -------------------------------
review_templates = {
    "phone": {
        5: "Điện thoại pin khỏe, màn hình đẹp, sạc nhanh. Tôi rất hài lòng!",
        4: "Điện thoại dùng ổn, camera khá tốt, chỉ pin hơi yếu.",
        3: "Máy dùng được, nhưng hiệu năng không nổi bật.",
        2: "Máy hay bị nóng và lag khi chơi game.",
        1: "Điện thoại lỗi, sạc không vào, thật sự thất vọng."
    },
    "laptop": {
        5: "Laptop chạy nhanh, pin tốt, rất phù hợp cho công việc.",
        4: "Máy chạy mượt, nhưng bàn phím hơi cứng.",
        3: "Hiệu năng tạm ổn, quạt kêu hơi to.",
        2: "Máy bị đơ khi mở nhiều ứng dụng.",
        1: "Laptop lỗi main, không khởi động được."
    },
    "clothes": {
        5: "Áo đẹp, vải mềm, mặc thoải mái. Tôi rất ưng ý!",
        4: "Chất liệu tốt, màu đúng hình, chỉ hơi rộng chút.",
        3: "Áo bình thường, không đặc sắc.",
        2: "Vải thô và dễ nhăn, form không đẹp.",
        1: "Áo rách khi nhận hàng, rất tệ."
    }
}

# -------------------------------
# 3. Sinh dữ liệu ngẫu nhiên
# -------------------------------
item_counter = 1
for user in user_ids:
    for _ in range(REVIEWS_PER_USER):
        # Chọn danh mục ngẫu nhiên
        category = random.choice(CATEGORIES)
        item_id = f"{category.capitalize()}_{item_counter:04d}"
        item_counter += 1

        # Chọn rating ngẫu nhiên và review tương ứng
        rating = np.random.randint(1, 6)
        review = review_templates[category][rating]

        data.append([user, item_id, rating, review, category])

# -------------------------------
# 4. Tạo DataFrame & lưu CSV
# -------------------------------
df = pd.DataFrame(data, columns=["UserID", "ItemID", "Rating", "Review_VN", "Category"])

file_name = "data-ecom_quangnv.csv"
df.to_csv(file_name, index=False, encoding="utf-8-sig")

print(f"✅ Đã tạo dữ liệu e-commerce: {file_name}")
print(df.head())
print("\nThống kê số lượng theo danh mục:")
print(df["Category"].value_counts())
```
