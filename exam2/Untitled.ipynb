```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense
import warnings

# Bỏ qua các cảnh báo trong quá trình chạy
warnings.filterwarnings('ignore')

# --- KHAI BÁO CẤU HÌNH DỮ LIỆU ---
FILE_PATH = 'weather1.csv'
# *** CẬP NHẬT: Khôi phục danh sách thành phố mục tiêu theo yêu cầu ban đầu ***
TARGET_CITIES = ['Ha Noi', 'Hoa Binh', 'Ho Chi Minh City']
TARGET_COL = 'max' # Đặt mục tiêu dự đoán là Nhiệt độ Tối đa ('max')

# --- HÀM TẠO DỮ LI LIỆU GIẢ LẬP (MOCK DATA) - DỰ PHÒNG ---
def create_mock_data(city_name, days=365, target_col='max'):
    """Tạo dữ liệu thời tiết giả lập cho một thành phố."""
    np.random.seed(42)
    dates = pd.date_range(start='2020-01-01', periods=days, freq='D')
    
    # Tạo chuỗi nhiệt độ có tính chu kỳ và nhiễu ngẫu nhiên
    temp_base = 25 + 5 * np.sin(np.linspace(0, 2 * np.pi * 2, days))
    temp_noise = np.random.randn(days) * 1.5
    
    # Điều chỉnh nhiệt độ cơ sở theo thành phố (giả lập)
    if 'Ha Noi' in city_name:
        temp = temp_base - 2 + temp_noise
    elif 'Ho Chi Minh' in city_name:
        temp = temp_base + 3 + temp_noise
    else: # Hoa Binh (giả định)
        temp = temp_base + 1 + temp_noise

    df = pd.DataFrame({
        'Date': dates,
        'City': city_name,
        target_col: temp,  # Sử dụng TARGET_COL ('max')
        'Humidity': np.random.randint(60, 95, days),
        'WindSpeed': np.random.rand(days) * 10
    })
    return df

# --- BƯỚC 1: TẢI VÀ CHUẨN BỊ DỮ LIỆU THỰC TẾ (ƯU TIÊN) HOẶC DÙNG GIẢ LẬP (DỰ PHÒNG) ---
weather_df = None
try:
    # 1. Tải dữ liệu thực tế
    df_raw = pd.read_csv(FILE_PATH)
    
    # --- THÊM: In 5 dòng đầu của dữ liệu thô để kiểm tra cấu trúc ---
    print(f"--- 5 DÒNG ĐẦU CỦA DỮ LIỆU THÔ TỪ '{FILE_PATH}' ---")
    print(df_raw.head())
    print("-" * 50)
    
    # 2. Tiền xử lý tên cột và ngày tháng
    if 'province' in df_raw.columns:
        df_raw = df_raw.rename(columns={'province': 'City'})
    if 'date' in df_raw.columns:
        # Sử dụng format '1/1/2009' theo mẫu
        # Cần xử lý cả trường hợp cột 'date' đã được tải thành datetime object (nếu format không chuẩn)
        try:
             df_raw['Date'] = pd.to_datetime(df_raw['date'], format='%Y-%m-%d', errors='coerce')
        except ValueError:
             df_raw['Date'] = pd.to_datetime(df_raw['date'], errors='coerce')

    # 3. Lọc và chuẩn bị dữ liệu
    
    # *** CẬP NHẬT: Chuẩn hóa tên thành phố để khớp với danh sách TARGET_CITIES không dấu ***
    # Bản đồ ánh xạ toàn diện hơn để xử lý các biến thể có/không dấu và khoảng trắng
    city_normalization_map = {
        'Hà Nội': 'Ha Noi',
        'Ha Noi': 'Ha Noi', # Xử lý nếu dữ liệu đã không dấu
        'Hoa Binh': 'Hoa Binh',
        'Hòa Bình': 'Hoa Binh',
        'Hồ Chí Minh': 'Ho Chi Minh City',
        'Ho Chi Minh': 'Ho Chi Minh City',
        'Ho Chi Minh City': 'Ho Chi Minh City',
        'TP HCM': 'Ho Chi Minh City',
        # Các tên miền Nam khác (giữ lại phòng trường hợp file có cả 2 miền)
        'Bạc Liêu': 'Bac Lieu',
        'Bac Lieu': 'Bac Lieu',
        'Sóc Trăng': 'Soc Trang',
        'Soc Trang': 'Soc Trang',
    }
    
    # Áp dụng ánh xạ để chuyển tên thành phố có dấu thành không dấu (theo TARGET_CITIES)
    # Sử dụng str.strip().title() để chuẩn hóa chuỗi tên tỉnh trước khi ánh xạ
    df_raw['City'] = df_raw['City'].astype(str).str.strip()
    df_raw['City'] = df_raw['City'].apply(
        lambda x: city_normalization_map.get(x, x)
    )
    
    # Lọc theo danh sách TARGET_CITIES: ['Ha Noi', 'Hoa Binh', 'Ho Chi Minh City']
    weather_df = df_raw[df_raw['City'].isin(TARGET_CITIES)].copy()
    weather_df = weather_df.sort_values(by=['City', 'Date']).reset_index(drop=True)

    # Đảm bảo cột mục tiêu là số và xử lý thiếu (loại bỏ hàng có NaNs trong cột mục tiêu)
    weather_df[TARGET_COL] = pd.to_numeric(weather_df[TARGET_COL], errors='coerce')
    weather_df.dropna(subset=['City', 'Date', TARGET_COL], inplace=True) 

    if len(weather_df) == 0:
        # Lỗi này giờ sẽ chỉ xảy ra nếu không có dữ liệu nào khớp sau khi chuẩn hóa
        raise ValueError(f"Không tìm thấy dữ liệu hợp lệ cho {TARGET_CITIES} trong file.")
    
    print(f"--- Đã tải dữ liệu thực tế từ '{FILE_PATH}' ---")

except (FileNotFoundError, ValueError, KeyError) as e:
    print(f"LỖI TẢI DỮ LIỆU THỰC TẾ ({e}). Chuyển sang sử dụng DỮ LIỆU GIẢ LẬP.")
    
    # --- TẠO DỮ LIỆU GIẢ LẬP (DỰ PHÒNG) Dựa trên TARGET_CITIES MỚI ---
    Hanoi_df = create_mock_data('Ha Noi', target_col=TARGET_COL)
    HoaBinh_df = create_mock_data('Hoa Binh', target_col=TARGET_COL)
    HCMC_df = create_mock_data('Ho Chi Minh City', target_col=TARGET_COL)
    weather_df = pd.concat([Hanoi_df, HoaBinh_df, HCMC_df], ignore_index=True)
    
    print("--- Dữ liệu giả lập đã được tạo và tải (Dự phòng) ---")

print(f"--- 5 DÒNG ĐẦU CỦA DỮ LIỆU ĐÃ LỌC VÀ CHUẨN BỊ (Chỉ hiển thị City và {TARGET_COL}) ---")
print(weather_df[['City', TARGET_COL]].head())
print("\n" + "="*50 + "\n")


# --- PHA 1: TIỀN XỬ LÝ DỮ LIỆU & TẠO TRÌNH TỰ ---

def preprocess_and_sequence(city_data, sequence_length=15, target_col='max'):
    """
    Thực hiện tỷ lệ hóa và tạo chuỗi thời gian (X, Y) cho mô hình RNN/LSTM.
    """
    print(f"Bắt đầu tiền xử lý cho {city_data['City'].iloc[0]}...")
    
    # Chọn đặc trưng cần dùng (chỉ dùng nhiệt độ cho ví dụ đơn giản này)
    data = city_data[[target_col]].values.astype('float32')
    
    # Tỷ lệ hóa dữ liệu (Scaling)
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    
    # Tạo tập X (đầu vào - sequence_length ngày trước) và Y (đầu ra - ngày tiếp theo)
    X, Y = [], []
    for i in range(len(scaled_data) - sequence_length):
        X.append(scaled_data[i:i + sequence_length, 0])
        Y.append(scaled_data[i + sequence_length, 0])
    
    # Chuyển đổi sang mảng numpy và định hình lại cho RNN/LSTM (samples, time_steps, features)
    X = np.array(X)
    Y = np.array(Y)
    # RNN/LSTM yêu cầu định dạng (samples, time_steps, features)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    
    # Chia tập huấn luyện (80%) và kiểm tra (20%)
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, shuffle=False
    )
    
    print(f"Kích thước X_train: {X_train.shape}, Kích thước X_test: {X_test.shape}")
    
    return X_train, X_test, Y_train, Y_test, scaler


# --- PHA 2: XÂY DỰNG MÔ HÌNH ---

def build_rnn_model(input_shape):
    """Xây dựng mô hình SimpleRNN."""
    model = Sequential([
        SimpleRNN(units=50, activation='relu', input_shape=input_shape, return_sequences=False),
        Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_lstm_model(input_shape):
    """Xây dựng mô hình LSTM."""
    model = Sequential([
        LSTM(units=50, activation='relu', input_shape=input_shape, return_sequences=False),
        Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model


# --- PHA 3 & 5: HUẤN LUYỆN, ĐÁNH GIÁ VÀ SO SÁNH ---

def train_evaluate_plot(city_name, X_train, X_test, Y_train, Y_test, scaler, model_type, target_col):
    """Huấn luyện mô hình, đánh giá và vẽ biểu đồ kết quả."""
    
    input_shape = (X_train.shape[1], X_train.shape[2])
    
    if model_type == 'RNN':
        model = build_rnn_model(input_shape)
    elif model_type == 'LSTM':
        model = build_lstm_model(input_shape)
    else:
        raise ValueError("Loại mô hình không hợp lệ.")

    print(f"\n--- Bắt đầu huấn luyện mô hình {model_type} cho {city_name} ---")
    
    # Huấn luyện mô hình
    history = model.fit(
        X_train, Y_train,
        epochs=50, 
        batch_size=32, 
        validation_split=0.1, 
        verbose=0 # Đặt verbose=1 để xem tiến trình huấn luyện chi tiết
    )
    
    # Dự đoán trên tập kiểm tra
    Y_pred_scaled = model.predict(X_test)
    
    # Đảo tỷ lệ (Inverse transform) để đưa về giá trị nhiệt độ thực tế
    Y_test_actual = scaler.inverse_transform(Y_test.reshape(-1, 1))
    Y_pred_actual = scaler.inverse_transform(Y_pred_scaled)
    
    # Đánh giá các độ đo
    mae = mean_absolute_error(Y_test_actual, Y_pred_actual)
    mse = mean_squared_error(Y_test_actual, Y_pred_actual)
    
    print(f"Kết quả {model_type} cho {city_name}:")
    print(f"  MAE (Mean Absolute Error) cho '{target_col}': {mae:.3f} °C")
    print(f"  MSE (Mean Squared Error) cho '{target_col}': {mse:.3f}")
    
    # --- TRỰC QUAN HÓA (ĐIỂM 4) ---
    plt.figure(figsize=(14, 6))
    
    # Biểu đồ so sánh giá trị thực tế và dự đoán
    plt.plot(Y_test_actual, label='Giá trị Thực tế (Actual)', color='blue')
    plt.plot(Y_pred_actual, label='Giá trị Dự đoán (Predicted)', color='red', linestyle='--')
    
    plt.title(f'Dự báo {target_col}: {city_name} - Mô hình {model_type}')
    plt.xlabel('Ngày (trên tập kiểm tra)')
    plt.ylabel(f'{target_col} (°C)')
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.show()
    
    return {'model_type': model_type, 'city': city_name, 'MAE': mae, 'MSE': mse}


# --- CHẠY CHƯƠNG TRÌNH CHÍNH ---

if __name__ == '__main__':
    
    cities = TARGET_CITIES # ['Ha Noi', 'Hoa Binh', 'Ho Chi Minh City']
    results = []

    # Kiểm tra xem dữ liệu có được tải thành công không
    if weather_df is None or len(weather_df) == 0:
        print("\nKHÔNG THỂ TIẾN HÀNH HUẤN LUYỆN DO DỮ LIỆU RỖNG HOẶC KHÔNG HỢP LỆ.")
        # Thoát nếu không có dữ liệu để chạy
        exit()
    
    for city in cities:
        # Lọc dữ liệu theo thành phố (PHA 1)
        city_df = weather_df[weather_df['City'] == city].copy()
        
        # Chỉ chạy nếu có đủ dữ liệu cho thành phố này (ít nhất sequence_length + 1)
        if len(city_df) < 16:
             print(f"Bỏ qua {city} do không đủ dữ liệu ({len(city_df)} bản ghi).")
             continue

        # Tiền xử lý và tạo chuỗi (PHA 1)
        # Truyền TARGET_COL vào hàm preprocess_and_sequence
        X_train, X_test, Y_train, Y_test, scaler = preprocess_and_sequence(
            city_df, target_col=TARGET_COL
        )
        
        # --- Chạy mô hình RNN ---
        result_rnn = train_evaluate_plot(
            city, X_train, X_test, Y_train, Y_test, scaler, 'RNN', TARGET_COL
        )
        results.append(result_rnn)
        
        # --- Chạy mô hình LSTM ---
        result_lstm = train_evaluate_plot(
            city, X_train, X_test, Y_train, Y_test, scaler, 'LSTM', TARGET_COL
        )
        results.append(result_lstm)

    # --- SO SÁNH TỔNG QUAN CÁC MÔ HÌNH (ĐIỂM 4 & 5) ---
    if results:
        results_df = pd.DataFrame(results)
        
        print("\n" + "="*50)
        print(f"BẢNG TỔNG KẾT VÀ SO SÁNH HIỆU SUẤT MÔ HÌNH (Dự đoán cột '{TARGET_COL}')")
        print("="*50)
        print(results_df.to_markdown(index=False))
        
        # Biểu đồ cột so sánh MAE
        plt.figure(figsize=(12, 6))
        
        # Tạo cột kết hợp City và Model Type để vẽ
        results_df['Label'] = results_df['city'] + ' (' + results_df['model_type'] + ')'
        
        # Chọn màu sắc khác nhau để dễ phân biệt
        colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974', '#64B5CD']
        num_models = len(results_df)
        plt.bar(
            results_df['Label'], 
            results_df['MAE'], 
            color=[colors[i % len(colors)] for i in range(num_models)]
        )
        
        plt.title(f'So sánh MAE (Lỗi Tuyệt đối Trung bình) cho cột "{TARGET_COL}"')
        plt.xlabel('Mô hình và Thành phố')
        plt.ylabel('MAE (°C)')
        plt.xticks(rotation=20, ha='right')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    else:
        print("\nKhông có mô hình nào được huấn luyện thành công.")
```
