```python
# 1.7 - BMI Classification ML
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# -----------------------
# 1. Generate 500 persons
# -----------------------
np.random.seed(42)
n = 500
heights = np.random.uniform(1.4, 2.0, n)   # chiu cao (m)
weights = np.random.uniform(40, 100, n)    # cn nng (kg)
bmi = weights / (heights ** 2)

# Label: 0=Underweight, 1=Normal, 2=Overweight
labels = np.where(bmi < 18.5, 0, np.where(bmi > 25, 2, 1))

# DataFrame
df = pd.DataFrame({
    "Height": heights,
    "Weight": weights,
    "BMI": bmi,
    "Label": labels
})
print(df.head())

# -----------------------
# 2. Train/Test split
# -----------------------
X = df[["Height", "Weight"]]
y = df["Label"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# -----------------------
# 3. Apply 3 ML techniques
# -----------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(n_neighbors=5)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=["Underweight","Normal","Overweight"])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

# -----------------------
# 4. Compare accuracy
# -----------------------
plt.bar(results.keys(), results.values(), color=["orange", "green", "blue"])
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.ylim(0, 1)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# -------------------------
# 1. Sinh d liu 500 sinh vin
# -------------------------
np.random.seed(42)
n = 500

toan = np.random.uniform(4, 10, n)         # im Ton (4-10)
python = np.random.uniform(4, 10, n)       # im Lp trnh Python
csdl = np.random.uniform(4, 10, n)         # im CSDL

# im PHTTTM ph thuc vo 3 mn trn
# Gi nh cng thc: 40% Python + 30% Ton + 30% CSDL + noise
noise = np.random.normal(0, 0.5, n)        # nhiu
phtttm = 0.4*python + 0.3*toan + 0.3*csdl + noise
phtttm = np.clip(phtttm, 0, 10)            # gii hn 010

# DataFrame
df = pd.DataFrame({
    "Ton": toan,
    "Python": python,
    "CSDL": csdl,
    "PHTTTM": phtttm
})
print(df.head())

# -------------------------
# 2. Train/Test split
# -------------------------
X = df[["Ton", "Python", "CSDL"]]
y = df["PHTTTM"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# -------------------------
# 3. Hun luyn 3 m hnh
# -------------------------
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = (mse, r2)

    # V biu  so snh d on vs thc t (10 sinh vin u)
    plt.figure(figsize=(6,4))
    plt.plot(range(10), y_test[:10], marker='o', label="Thc t")
    plt.plot(range(10), y_pred[:10], marker='s', label="D on")
    plt.title(f"So snh d on - {name}")
    plt.xlabel("SV")
    plt.ylabel("im PHTTTM")
    plt.legend()
    plt.show()

# -------------------------
# 4. So snh MSE v R
# -------------------------
models_name = list(results.keys())
mse_vals = [results[m][0] for m in models_name]
r2_vals = [results[m][1] for m in models_name]

plt.figure(figsize=(8,4))
plt.bar(models_name, mse_vals, color="orange")
plt.ylabel("MSE (cng thp cng tt)")
plt.title("So snh MSE cc m hnh")
plt.show()

plt.figure(figsize=(8,4))
plt.bar(models_name, r2_vals, color="green")
plt.ylabel("R (cng gn 1 cng tt)")
plt.title("So snh R cc m hnh")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# -------------------------
# 1. Sinh d liu 500 sinh vin
# -------------------------
np.random.seed(42)
n = 500

toan = np.random.uniform(4, 10, n)         
python = np.random.uniform(4, 10, n)       
csdl = np.random.uniform(4, 10, n)        

# Cng thc to im PHTTTM
noise = np.random.normal(0, 0.5, n)        
phtttm = 0.4*python + 0.3*toan + 0.3*csdl + noise
phtttm = np.clip(phtttm, 0, 10)            

# -------------------------
# 2. Chuyn thnh nhn hc lc
# -------------------------
def classify(score):
    if score < 6.5:
        return 0   # Trung bnh
    elif score < 8:
        return 1   # Kh
    else:
        return 2   # Gii

labels = np.array([classify(s) for s in phtttm])

df = pd.DataFrame({
    "Ton": toan,
    "Python": python,
    "CSDL": csdl,
    "PHTTTM": phtttm,
    "Hc_lc": labels
})
print(df.head())

# -------------------------
# 3. Train/Test split
# -------------------------
X = df[["Ton", "Python", "CSDL"]]
y = df["Hc_lc"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# -------------------------
# 4. Logistic Regression
# -------------------------
model = LogisticRegression(max_iter=1000, multi_class="multinomial")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# -------------------------
# 5. Evaluation
# -------------------------
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=["Trung bnh", "Kh", "Gii"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
# -------------------------
# 6. D on cho d liu mi
# -------------------------
import tkinter as tk
from tkinter import messagebox

# -------------------------
# 6. To giao din  nhp v d on
# -------------------------
def predict_from_gui():
    try:
        # Ly gi tr t cc  nhp liu
        toan_new = float(entry_toan.get())
        python_new = float(entry_python.get())
        csdl_new = float(entry_csdl.get())

        # To DataFrame v d on
        new_data = pd.DataFrame({
            "Ton": [toan_new],
            "Python": [python_new],
            "CSDL": [csdl_new]
        })
        predicted_label = model.predict(new_data)[0]

        # nh x nhn s sang nhn ch
        labels_map = {0: "Trung bnh", 1: "Kh", 2: "Gii"}
        predicted_class = labels_map[predicted_label]

        # Hin th kt qu trong mt hp thoi
        messagebox.showinfo("D on hc lc", f"Hc lc ca bn l: {predicted_class}")
    except ValueError:
        messagebox.showerror("Li", "Vui lng nhp s hp l cho tt c cc mn.")

# To ca s chnh
root = tk.Tk()
root.title("D on Hc Lc")
root.geometry("400x250")
root.configure(bg="#f0f0f0")

# To khung cha cc widget
frame = tk.Frame(root, padx=20, pady=20, bg="#f0f0f0")
frame.pack(expand=True)

# Tiu 
title_label = tk.Label(frame, text="Nhp im ca bn", font=("Helvetica", 16, "bold"), bg="#f0f0f0")
title_label.grid(row=0, column=0, columnspan=2, pady=(0, 20))

# Cc label v  nhp liu
tk.Label(frame, text="im mn Ton:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=1, column=0, sticky="w", pady=5)
entry_toan = tk.Entry(frame, font=("Helvetica", 12))
entry_toan.grid(row=1, column=1, pady=5)

tk.Label(frame, text="im mn Python:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=2, column=0, sticky="w", pady=5)
entry_python = tk.Entry(frame, font=("Helvetica", 12))
entry_python.grid(row=2, column=1, pady=5)

tk.Label(frame, text="im mn CSDL:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=3, column=0, sticky="w", pady=5)
entry_csdl = tk.Entry(frame, font=("Helvetica", 12))
entry_csdl.grid(row=3, column=1, pady=5)

# Nt d on
predict_button = tk.Button(frame, text="D on", font=("Helvetica", 12, "bold"), bg="#4CAF50", fg="white", command=predict_from_gui)
predict_button.grid(row=4, column=0, columnspan=2, pady=20, ipadx=20, ipady=5)

# Chy vng lp chnh ca Tkinter
root.mainloop()
```
